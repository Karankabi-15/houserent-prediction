# EDA

## Bar Plot for number of house in each City which is available for rent

```python
    sns.set_context("poster", font_scale=.8)
    plt.figure(figsize=(20, 6))
    ax = rent_id["City"].value_counts().plot(kind='bar', color='crimson', rot=0)
    
    for p in ax.patches:
        ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() - 100), 
                    ha='center', va='bottom', color='white')
    
    plt.show()
```

## 'Bar Plot for number of house in each Furnishing Status which is available for rent

```python
    sns.set_context("poster", font_scale=.8)
    plt.figure(figsize=(20, 6))
    ax = rent_id["Furnishing Status"].value_counts().plot(kind='bar', color='crimson', rot=0)
    
    for p in ax.patches:
        ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() - 100), 
                    ha='center', va='bottom', color='white')
    
    plt.show()
```

```python
    sns.set_context("poster", font_scale=.8)
    plt.figure(figsize=(20, 6))
    ax = rent_id["Floor"].value_counts().plot(kind='bar', color='crimson', rot=0)
    
    for p in ax.patches:
        ax.annotate(int(p.get_height()), (p.get_x() + 0.25, p.get_height() - 100), 
                    ha='center', va='bottom', color='white')
    
    plt.show()
```

```python
plt.figure(figsize=(20, 6))
counts=rent_id["City"].value_counts()
explode=(0,0,0,0.1,0,0)
colors=["#FF1E00","#FF4500","#FF6347","#FF5733","#FF3300","#FF6F61"]
counts.plot(kind="pie",colors=colors,explode=explode,autopct='%1.1f%%')
plt.axis('equal')
plt.legend(labels=counts.index,loc='best')
plt.show()
```
